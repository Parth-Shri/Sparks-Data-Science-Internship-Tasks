## Visualization

```python
plt.figure(figsize = (6,2))
sns.pairplot(unemployment)
plt.subplots_adjust(hspace = 0.5)
```

```python
sns.pairplot(unemployment_2011)
```

```python
plt.figure(figsize = (8,6))
plt.bar(unemployment['Region'], unemployment[' Estimated Unemployment Rate (%)'])
plt.title("Unemployment rate in different states")
plt.xlabel("States")
plt.ylabel("Unemployment Rate")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.bar(unemployment['Region'], unemployment[' Estimated Employed'])
plt.title("Employment rate in different states")
plt.xticks(rotation = 90)
plt.xlabel("States")
plt.ylabel("Employment Rate")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.lineplot(x = ' Date', y = ' Estimated Unemployment Rate (%)', data = unemployment, hue = 'Area')
plt.xticks(rotation = 90)
plt.title("Unemployment according to date")
plt.show()
```

```python
plt.scatter(x = unemployment['Region'], y = unemployment[' Estimated Labour Participation Rate (%)'])
plt.xticks(rotation = 90)
plt.plot()
```

## For second dataset

```python
plt.figure(figsize = (10,6))
sns.barplot(x = unemployment_2011['Region'], y = unemployment_2011[' Estimated Unemployment Rate (%)'], data = unemployment_2011, errorbar = None)
plt.xticks(rotation = 90)
plt.title("States v/s Unemployment in 2011")
plt.xlabel("States")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.barplot(x = 'Region', y = ' Estimated Employed', data = unemployment_2011, errorbar = None)
plt.title("States v/s Employment in 2011")
plt.xticks(rotation = 90)
plt.xlabel("States")
plt.ylabel("Employment Rate")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.lineplot(x = ' Date', y = ' Estimated Unemployment Rate (%)', data = unemployment_2011)
plt.xticks(rotation = 90)
plt.title("Unemployment in 2011 according to Date")
plt.show()
```

```python
sns.scatterplot(x = 'Region.1', y = ' Estimated Labour Participation Rate (%)', data = unemployment_2011)
plt.xticks(rotation = 90)
plt.show()
```
